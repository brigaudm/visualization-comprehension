```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
abs_errors = files.upload()
```

```python
abs_errors = pd.read_excel("cmt218-dve-2020-2024(1).xlsx", sheet_name="abs-errors")
```

```python
abs_errors
```

```python
melted_abs_errors = pd.melt(abs_errors, id_vars=['Team', 'Type', 'Attempt', 'Year'], var_name='Vis-Value', value_name='Error')
melted_abs_errors
```

```python
error_averages = melted_abs_errors[['Type', 'Error', 'Year']].groupby('Type').agg(['mean', 'count'])
error_averages.columns = error_averages.columns.droplevel(0)
error_averages.sort_values(by='mean', ascending=True)
```

```python
error_2024 = melted_abs_errors[melted_abs_errors['Year'] == '23-24']
error_2024
```

```python
sns.barplot(data=error_averages.sort_values(by='mean', ascending=False), x='Type', y='mean')
```
